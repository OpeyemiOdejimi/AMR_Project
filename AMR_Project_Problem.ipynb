# AI-Powered Antimicrobial Resistance (AMR) Surveillance and Decision Support System

## The Problem

**Antimicrobial Resistance (AMR)** is a silent killer in Nigeria's health system for a couple of reasons:

- Increase in mortality rate and hospital stay.
- Skyrocketing the costs of treatment for average-earning Nigerians or low-income citizens.
- Crippling Nigeria's epidermic preparedness e.t.c.

Generally speaking, quite a large number of Nigerians are ignorant on the concept of Antimicrobial Resistance. If you were to pick one person, from each of the 36 states in Nigeria and you ask them what AMR is, chances are less than 10 would be able to say anything. This lack of proper or adequate public sensitization has led to the silent spread of Antimicrobial resistance in the society.

To put it succinctly, AMR in Nigeria is a complex problem currently being motored by a plethora of healthcare gap/negligence, agricultural practices and socioeconomic factors. Unlike a sudden epidemic outbreak, AMR grows gradually when bacteria adapt to drugs that naturally would kill them. I will further discuss these factors that contribute to these problems below:

1. **Unregulated Access and Self-Medication:** According to research by experts, one of the key drivers of AMR in Nigeria is the ease at which antibiotics are accessible. In Nigeria, potent antibiotics like penicillin, amoxicillin, ampicillin etc are sold without prescriptions at pharmacies and by mostly unqualified personnel commonly referred to as "Chemists", or even street vendors. 

Another issue is the habit of not completing drug courses, this sometimes is due to high cost of such drugs and sheer negligence or irresponsibility on the parts of patients. Some of them abandon the drugs as soon as they feel better, this allows the bacteria to survive and develop resistance to the antibiotic in question. Nigerians also have the habit of taking antibiotics for every thing, for example most Nigerians will take antibiotics when they have a cold or flu. These infections are viral and they usually subside after about 14 days, these antibiotics go into the body and the resident bacterial develop resistance to them.

2. **Lack of Diagnostic Infrastructure:** Many health facilities in Nigeria lack basic infrastructure to perform culture and sensitivity tests. This leads to medical practitioners blindly prescribing drugs to patients, because doctors cannot confirm if an infection is viral, fungal or bacterial, they in turn prescribe broad-spectrum antibiotics just in case and with the hope that the patient is actually suffering from a bacterial infection.

3. **Misuse in Agriculture and Livestock:** The Agricultural sector in Nigeria is a major yet commonly overlooked culprit in AMR development in Nigeria. This is because farmers have the habit of frequently using antibiotics in poultry and cattle feed, not to treat sickness per say but to enhance rapid growth and on some occasions to prevent diseases outbreak in crowded conditions.

How does this affect humans? These resistant bacteria from these animals enter the human body via food chain consumption of meat, milk, egg. Water bodies contaminated by animal waste if consumed by humans is another vehicle for transmission.

4. **Hospital-Acquired Infections:** According to data-backed inferences, overcrowding and insufficient Infection Prevention and Control (IPC) measures in some hospitals mean that resistant strains of bacterial can be passed on from patient to patients on admission.

Other key drivers include Poor sanitation and water quality. ###  Current Gaps in AMR Surveillance in Nigeria

Nigeria has taken major steps in addressing AMR at large, national actions plans and surveillance coordinated by the Nigeria Centre for Disease Control (NCDC), Nigerian National Antimicrobial Stewardship Taskforce are some of such plans. However, it is important to note that the current surveillance plans are still being hindered by structural, technical, and operational gaps that limit the ability to provide timely, actionable data or intelligence for clinical and policy decision-making.

At the moment, AMR data in Nigeria is broken, and incoherent. Antimicrobial susceptibility data that is generated in Microbiological labs of tertiary hospitals, research centres across the nation is often isolated and stored in offline platforms.

Due to this fact, timely reporting to national surveillance platforms is inconsistent, and large volumes of laboratory data—particularly from private and sub-national facilities—never reach central repositories. As a result, national AMR reports are based on incomplete and geographically biased datasets that do not accurately reflect resistance patterns across the country.


While AMR data is collected for surveillance purposes, it is underutilized for operational decision-making. Clinicians often lack access to up-to-date, facility- or region-specific resistance patterns and are forced to rely on empirical prescribing or outdated *antibiograms*. Policymakers receive static reports that provide limited guidance for antimicrobial stewardship, procurement planning, or targeted interventions. The absence of early warning mechanisms and predictive analytics further weakens Nigeria’s capacity to anticipate and mitigate AMR risks.

These challenges are compounded by infrastructure and capacity constraints. Many health facilities operate in low-bandwidth environments with limited digital infrastructure, and there is a shortage of dedicated analytical capacity to transform raw AMR data into actionable intelligence. Furthermore, existing surveillance systems are urban-biased, with rural and underserved populations underrepresented, raising concerns about equity and the true national burden of AMR.

Collectively, these gaps mean that while AMR data exists in Nigeria, the health system lacks an integrated, intelligent mechanism to convert this data into timely insights that can inform clinical practice, public health response, and policy formulation. Without addressing these systemic weaknesses, AMR will continue to erode health outcomes and development gains.

There is therefore an urgent need for an integrated, scalable, and context-appropriate solution that strengthens Nigeria’s AMR surveillance architecture by enabling real-time data integration, predictive analysis, early warning, and feedback to clinicians and policymakers. Leveraging **artificial intelligence** and **machine learning** as an enabling layer—rather than a replacement for existing systems—offers a viable pathway to enhance surveillance effectiveness, improve antimicrobial stewardship, and support evidence-based decision-making.
